```python
def curso(a):
  import json 
  import requests
  from pprint import pprint
  response=  requests.get("https://fp.josedomingo.org/lmgs/u05/ej2.json")
  pendientes= json.loads(response.text)

  for i in pendientes:
    if i.get("ID")== a:
      b=i.get("Titulo")
      c= i.get("Profesorado")
      d=""
      for j in c:
        d+=j.get("NombreCompleto")
        d+=", "
      e = str(a)+" "+ str(b)+" Profesores: "+str(d)
      print(e)
```
