```python
import matplotlib.pyplot as plt
import random
imagen = []
for i in range(0, 200):
  fila = []
  for j in range(0, 200):
    if i<j:
      rojo = 255
      verde = 255
      azul = 255
    elif i==j:
      rojo = random.randint(0,255)
      verde = random.randint(0,255)
      azul = random.randint(0,255)
    else:
      rojo = 0
      verde = 0
      azul = 0
    pixel = [rojo, verde, azul]
    fila.append(pixel)
  imagen.append(fila)
plt.imshow(imagen)
```

```python
from skimage import exposure
from skimage import data
import matplotlib.pyplot as plt
gato = data.cat()
img_eq = exposure.equalize_hist(gato)
for eq in (gato, img_eq):
    plt.imshow(eq)
    plt.show()
```
